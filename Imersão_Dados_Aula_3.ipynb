```python
topBairros = dadosVendas['Bairro'].value_counts()[:10].index
```

```python
#dadosBairro.reset_index(inplace = True)
dadosBairro = dadosBairro.reset_index()
```

```python
dadosBairro.query('Bairro in @topBairros')
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Bairro', y='ValorM2Bairro', data=dadosBairro.query('Bairro in @topBairros'))
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dadosVendas.query('Bairro in @topBairros'), x="Bairro", y='ValorAnuncio')
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dadosVendas.query('Bairro in @topBairros & Metragem < 30000'), x="Bairro", y='Metragem')
ax.tick_params(axis='x', rotation=45)
plt.show()
```
